```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
rent_arousal = []
rent_valence = []

with open("results/rent_arousal.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        rent_arousal.append(float(row[0]))
        
with open("results/rent_valence.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        rent_valence.append(float(row[0]))

time1 = [i*50 for i in range(len(rent_arousal))]
time2 = [i*50 for i in range(len(rent_valence))]
m,b = np.polyfit(time1, rent_arousal, 1)
m2,b2 = np.polyfit(time2, rent_valence, 1)

plt.plot(time1, rent_arousal, label="arousal")
plt.plot(time1,np.array(time1) * m + b)
plt.plot(time2, rent_valence, label="valence")
plt.plot(time2,np.array(time2) * m2 + b2)
plt.title('Emotion Scores for Rent')
plt.xlabel('Time (sec)')
plt.ylabel('Score')
plt.legend()
plt.show()
```
