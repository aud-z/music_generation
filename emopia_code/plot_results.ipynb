```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
rent_arousal = []
rent_valence = []

with open("results/rent_arousal.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        rent_arousal.append(float(row[0]))
        
with open("results/rent_valence.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        rent_valence.append(float(row[0]))

time1 = [i*50 for i in range(len(rent_arousal))]
time2 = [i*50 for i in range(len(rent_valence))]
m,b = np.polyfit(time1, rent_arousal, 1)
m2,b2 = np.polyfit(time2, rent_valence, 1)

plt.plot(time1, rent_arousal, label="arousal")
plt.plot(time1,np.array(time1) * m + b)
plt.plot(time2, rent_valence, label="valence")
plt.plot(time2,np.array(time2) * m2 + b2)
plt.title('Emotion Scores for Rent')
plt.xlabel('Time (sec)')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
def convert_to_min(sec):
    m = (sec // 60) % 60
    h = (sec // 60) // 60
    s = sec % 60
    return (h, m, s)

# print([(i, convert_to_min(time1[i])) for i in range(len(time1))])

a1s1 = range(5, 14)
a1s2 = range(14, 27)
a1s3 = range(27, 34)
a1s4 = range(34, 37)
a1s5 = range(37, 48)
a2s1 = range(48, 53)
a2s2 = range(53, 82)
a3s1 = range(82, 94)
a3s2 = range(94, 112)
a3s3 = range(112, 117)
a3s4 = range(117, 128)
a4s1 = range(128, 129)
a4s3 = range(129, 134)
a4s4 = range(134, 138)
a4s5 = range(138, 147)
a4s7 = range(147, 155)
a5s1 = range(155, 173)
a5s2 = range(173, 190)

hamlet = [a1s1, a1s2, a1s3, a1s4, a1s5, a2s1, a2s2, a3s1, a3s2, a3s3, a3s4,
         a4s1, a4s3, a4s4, a4s5, a4s7, a5s1, a5s2]

avgs_arousal = []
avgs_valence = []
for scene in hamlet:
    total_arousal = 0
    total_valence = 0
    for img in scene:
        total_arousal += hamlet_arousal[img]
        total_valence += hamlet_valence[img]
    avgs_arousal.append(total_arousal / len(scene))
    avgs_valence.append(total_valence / len(scene))
print(avgs_arousal)
print(avgs_valence)

plt.plot(range(len(avgs_arousal)), avgs_arousal, label = "arousal")
plt.plot(range(len(avgs_valence)), avgs_valence, label = "valence")
plt.title('Emotion Scores for Hamlet Scenes')
plt.xlabel('Scene Number')
plt.ylabel('Score')
plt.legend()
plt.show()

hamlet_emotions = []
for i in range(len(avgs_arousal)):
    if avgs_arousal[i] < 0 and avgs_valence[i] < 0:
        hamlet_emotions.append(3)
    elif avgs_arousal[i] < 0:
        hamlet_emotions.append(4)
    elif avgs_valence[i] < 0:
        hamlet_emotions.append(2)
    else:
        hamlet_emotions.append(1)

print(hamlet_emotions)
```
